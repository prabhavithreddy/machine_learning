### 1. Importing the required libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### 2. Load the Image using the imread function of OpenCV

```python
image = cv2.imread('./RussianCar.jpg')
image.shape
```

### 3. Pre-process the Image 
    Create a function to convert the image using cv2.Color(img, cv2.COLOR_BGR2RGB) function
    Display the image

```python
def convert_image(image):
    #image = cv2.resize(image, (800,400))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = np.array(image, dtype='uint8')
    plt.imshow(image)
    return image

c = convert_image(image)
```

### 4. Create a function for a classifier using the HaarCascade_russian_plate_number.xml file
        Detect the Car License Plates using the Classifier
        Display the Results

```python
classifier = cv2.CascadeClassifier('./DL - MCT Data/haarcascade_russian_plate_number.xml')
converted_image = convert_image(image)
cars = classifier.detectMultiScale(converted_image, scaleFactor=1.05, minNeighbors=7)
for (x,y,w,h) in cars:
    cv2.rectangle(converted_image, (x,y), (x+w, y+h), (0,255,255), 2)
    #cv2.imshow('Cars', frame)
    plt.imshow(converted_image)
    #axarr[count//75,count%75].imshow(frame)
    plt.show()
```
