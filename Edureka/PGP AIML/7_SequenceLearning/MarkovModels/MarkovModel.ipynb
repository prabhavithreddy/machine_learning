```python
import pandas as pd
import numpy as np
```

```python
customers = [f'Cust_{i}' for i in np.arange(1,11)]
customers
```

```python
products = ['T-Shirts', 'Shoes', 'Denims']
products
```

```python
columns = ['Cust', 'Product']
months = ["Jan'18","Feb'18","Mar'18","Apr'18","May'18","Jun'18","Jul'18"]
columns.extend(months)
columns
```

```python
df = pd.DataFrame(columns=columns)
for customer in customers:
    for product in products:
        row = dict(zip(months, np.random.randint(0,2,7)))
        row['Cust'] = customer
        row['Product'] = product
        df = df.append(row, ignore_index=True)
df
```
