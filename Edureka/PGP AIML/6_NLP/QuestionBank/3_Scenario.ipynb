```python
df = pd.read_json('quotes.json')
df.head()
```

```python
df = df[['Quote']]
df.drop_duplicates(inplace=True)
df.head()
```

```python
from nltk.tokenize import word_tokenize
```

```python
from string import punctuation
import re
def frequency_table(sentence):
    sentence = sentence.lower()
    words = [word for word in word_tokenize(sentence)
            if word not in stop_words and word not in punctuation and re.match(r'^\w+$', word)]
    total_words.extend(words)
```

```python
from collections import Counter
total_words = []
df['Quote'].apply(frequency_table)
word_frequency = Counter(total_words)
word_frequency
```

```python
def score_sentences(sentence):
    sentence = sentence.lower()
    words = word_tokenize(sentence)
    word_count = 0
    frequency_count = 0
    for word in words:
        if word in word_frequency:
            word_count +=1
            frequency_count += word_frequency[word]
    if word_count == 0:
        return 0
    return frequency_count / word_count
```

```python
df['Score'] = df['Quote'].apply(score_sentences)
df.head()
```

```python
# average score
print(df['Score'].mean())
```
