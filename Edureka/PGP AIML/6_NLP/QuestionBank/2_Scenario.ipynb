```python
# Create a function frequency_table to calculate the Frequecy of Words present in the text
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
from nltk.stem import WordNetLemmatizer

stop_words = stopwords.words('English')
lemmatizer = WordNetLemmatizer()
```

```python
import re
def frequency_table(text):
    words = []
    text = text.lower()
    sentences = sent_tokenize(text)
    word_tokens = [word_tokenize(sentence) for sentence in sentences]
    for tokens in word_tokens:
        for word in tokens:
            if word not in stop_words and word not in punctuation and re.match('^\w+$', word):
                words.extend([lemmatizer.lemmatize(word)])
    global_words.extend(words)
    return " ".join(words)
```

```python
global_words = []
df['text'].apply(frequency_table)
```

```python
from collections import Counter
Counter(global_words)
```

```python
# Create a score function score_sentence that calculates the score for every sentence in the text
```

```python
def score_function(sentence):
    sentence = sentence.lower()
    words = [word for word in word_tokenize(sentence)
             if word not in stop_words and word not in punctuation and re.match('^\w+$', word)]
    #print(words)
    words_freq = Counter(words)
    sentence_freq = sum([words_freq[key] for key in words_freq.keys()])/len(sentence.split())
    return sentence_freq
```

```python
score_function('I am willing to kick some kick')
```

```python
df['Score'] = df['text'].apply(score_function)
df.head()
```

```python
# Create a function find_average_score to Find the Average Score
```

```python
def find_average_score():
    return df['Score'].mean()
```
