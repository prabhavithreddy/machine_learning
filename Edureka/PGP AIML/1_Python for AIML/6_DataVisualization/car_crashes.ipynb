```python
import numpy as nop
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv',index_col='abbrev')
df.head()
```

```python
plt.figure(figsize=(20,5))
sns.barplot(x=df.index,y=df.total)
```

```python
plt.figure(figsize=(20,10))
sns.relplot(x=df.index,y=df.total,kind='line',data=df)
```

```python
sns.relplot(data=df[['total','speeding','alcohol','not_distracted','no_previous']], kind='line',sizes=(20,10))
```
