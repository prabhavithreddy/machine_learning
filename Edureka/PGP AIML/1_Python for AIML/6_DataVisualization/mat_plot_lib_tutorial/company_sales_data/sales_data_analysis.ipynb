```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("company_sales_data.csv", index_col="month_number")
df.head()
```

#### Exercise 1: Read Total profit of all months and show it using a line plot

```python
plt.figure(figsize=(5,2))
plt.plot(df.total_profit, label='Total Profit',color='red', linestyle='dotted', linewidth=3, marker='o')
plt.xlabel = 'Month Number'
plt.ylabel = 'Sold units number'
plt.title = 'Company sales data of last year'
plt.legend(loc=4)
plt.show()
```

#### Exercise 3: Read all product sales data and show it  using a multiline plot

```python
plt.figure(figsize=(10,6))
for column in df.columns[:-2]:
    plt.plot(df[column], linewidth=3, marker='o')
plt.yticks = np.linspace(1000,18000, 18)
plt.ylabel='Sales units in number'
plt.xlabel='Month Number'
plt.legend()
plt.show()
```

#### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.figure(figsize=(12,6))
plt.scatter(x=df.index,y='toothpaste', data=df,label="Tooth paste sales data", linewidth=1)
plt.xlabel="Month Number"
plt.ylabel="Number of Units Sold"
plt.xticks = np.linspace(1,12,12)
plt.grid(True, linewidth=1, linestyle='--')
plt.legend(loc='upper left')
plt.show()
```

#### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
#plt.bar(x=[1],y=df['facecream'],height=1)
plt.bar([x for x in df.index.to_list()], df['facecream'].to_list(), label="Face cream sales data", width=0.25)
plt.bar([x+0.25 for x in df.index.to_list()], df['facewash'].to_list(), label="Face wash sales data", width=-0.25)
plt.xlabel="Month Number"
plt.ylabel="Sales units in number"
plt.legend()
plt.show()
```
