```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
v1 = np.random.randint(0,11, 2)
v1
```

```python
v2 = np.random.randint(0,11,2)
v2
```

```python
plt.plot(v1, label = 'v1')
plt.plot(v2, label = 'v2')
plt.plot(v1+v2, label = 'v1+v2')
plt.plot(v1*4, label='v1*4')
plt.plot(v2/2, label='v2/2')
plt.legend()
plt.show()
```
