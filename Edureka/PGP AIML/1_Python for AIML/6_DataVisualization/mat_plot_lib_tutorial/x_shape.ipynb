```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
smily_array = np.arange(-9,10)
smily_array
```

```python
plt.plot(smily_array, smily_array**2, label= "Smily")
plt.plot(smily_array, smily_array**2/2, label= "Teeth")
plt.legend()
plt.show()
```

```python
x_array = np.arange(0, 11)
x_array
```

```python
y_array = x_array
y_array
```

```python
plt.plot(x_array, y_array)
plt.plot(x_array, y_array[::-1])
plt.show()
```
