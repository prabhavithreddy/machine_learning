```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
```

```python
cl = .94
n = 200
d = 12

p = 12/200
print('Probability of defect is {}'.format(p))
```

```python
print(f'Success: {n*p}')
print(f'Failure: {n*(1-p)}')
```

```python
sd = np.sqrt((p*(1-p))/n)
print(f'SD: {sd}')
```

```python
se = sd/np.sqrt(n)
print(f'Standard Error: {se}')
```

```python
ci = 2*se
print(f'CI: {ci}')
```

```python
print(f"Margin of error: {p-ci}, {p+ci}")
```

```python
from scipy.stats import norm
norm.cdf(p+ci)
```

```python
def t_conf_interval(array, t):
    sample = np.array(array)
    df = len(sample)-1
    mean = np.round(np.mean(sample),2)
    print(f'df: {df}, mean: {mean}, std: {np.std(sample)}')
    me = t*(np.std(sample)/np.sqrt(5))
    print(f"Margin of error: {me}")
    print(f"[{mean-me},{mean+me}]")
```

```python
t_conf_interval([10,7,17,5,-2,20], 2.571)
```
