```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import sidetable as stb
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Fifa.csv', index_col=[0])
df.head()
```

```python
df.info()
```

### Question-1: How does a player's potential affect with his age? Visualize the same and write your inference on the same

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Potential', y='Age')
plt.xlabel('Potential')
plt.ylabel('Age')
plt.title('Potential vs Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.jointplot(data=df,x='Age', y='Potential')
plt.xlabel('Age')
plt.ylabel('Potential')
plt.title('Potential vs Age')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.displot(data=df,x='Potential', y='Age')
plt.xlabel('Potential')
plt.ylabel('Age')
plt.title('Potential vs Age')
plt.show()
```

### Question-2: Visualize the sprint speed of players as their age progresses. Write your observation

```python
sns.scatterplot(data=df,x='Age',y='SprintSpeed')
plt.xlabel('Age')
plt.ylabel('Speed')
plt.title('Age vs Speed')
plt.show()
```

```python
sns.displot(data=df,x='Age',y='SprintSpeed')
plt.xlabel('Age')
plt.ylabel('Speed')
plt.title('Age vs Speed')
plt.show()
```

```python
sns.lmplot(data=df,x='Age',y='SprintSpeed', lowess=True, 
           scatter_kws={'alpha':0.01, 's':3,'color':'blue'}, 
           line_kws={'color':'green'})
plt.xlabel('Age')
plt.ylabel('Speed')
plt.title('Age vs Speed')
plt.show()
```

### Question-3: Visualize to find whether the players are right or left footed

```python
sns.countplot(df['Preferred Foot'])
plt.show()
```

```python
sns.lmplot(data=df,x='BallControl', y='Dribbling', col='Preferred Foot', 
            scatter_kws ={'alpha':0.1, 'color':'yellow'}, line_kws={'color':'black'})
plt.show()
```

### Question-4: Compare the Overall Scores and age with respect to player's Preferred foot

```python
sns.scatterplot(data=df, x='Age', y='Overall', hue='Preferred Foot')
plt.show()
```

```python
sns.lmplot(data=df, x='Age', y='Overall', col='Preferred Foot')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (20,7)
plt.style.use('seaborn-dark-palette')

sns.boxenplot(data=df, x='Overall', y='Age', hue='Preferred Foot', palette='rocket')
plt.title('Comparison of overall scores and age wrt Preferred Foot', fontsize=20)
plt.show()
```
