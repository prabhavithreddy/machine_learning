```python
df = pd.read_csv('..\Data\Car_sales.csv')
df.head()
```

### Bivariate analysis
    categorical vs categorical -- cross tab
    categorical vs numerical --bar plot

```python
df.stb.missing().head(3)
```

```python
df['__year_resale_value'].fillna(df['__year_resale_value'].median(), inplace=True)
```

```python
df['__year_resale_value'].plot(kind='box')
plt.show()
```

```python
pd.crosstab(df['Manufacturer'],df['Vehicle_type'],normalize='columns').head()
```

```python
pd.crosstab(df['Manufacturer'],df['Model'],normalize='columns').head()
```

```python
pd.crosstab(df['Model'],df['Manufacturer'],normalize='columns').head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
for col in df.select_dtypes(exclude='object').columns:
    sns.scatterplot(col, 'Sales_in_thousands', hue='Vehicle_type',data=df)
    plt.title(f"{col} vs Sales_in_thousands")
    plt.show()
```
